**FACE RECOGNITION**





```python
import face_recognition



import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from matplotlib.patches import Circle
import numpy as np
import cv2 #opencv

%matplotlib inline
```

```python
# create a database

image = cv2.imread("dany.jpg")
dany = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)


image = cv2.imread("george-clooney.jpg")
clooney = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

image = cv2.imread("leonardo di capirio.jpg")
leonardo = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

image = cv2.imread('jon-snow.jpg')
jon = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)


```

```python
dany_encoding = face_recognition.face_encodings(dany)[0]
clooney_encoding = face_recognition.face_encodings(clooney)[0]
leonardo_encoding =  face_recognition.face_encodings(leonardo)[0]
jon_encoding =  face_recognition.face_encodings(jon)[0]


known_face_encodings= [dany_encoding,clooney_encoding,leonardo_encoding,jon_encoding]
```

```python
# load the image we want to check

image = cv2.imread('jon dany.jpg')

unknown_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(unknown_image)
```

```python
unknown_face_encodings = face_recognition.face_encodings(unknown_image)
```

```python
from scipy.spatial import distance

for unknown_face_encoding in unknown_face_encodings:

    results=[]

    for known_face_encoding in known_face_encodings:
        d = distance.euclidean(known_face_encoding, unknown_face_encoding)
        
        results.append(d)

    threshold = 0.6
    results = np.array(results)<= threshold

    name = 'unknown'

    if results[0]:
        name = 'Mother of dragon'
    elif results[1]:
        name = 'George Clooney'
    elif results[2]:
        name =  'Leonardo Dicaprio'
    elif results[3]:
        name = 'Jon snow'
  

    print(f'Found {name} in the photo!')
```
